```python
import pandas as pd

test = pd.read_csv("../data/bicikelj_test.csv")

all_predictions = []

i = 0
while i < len(test):
    section = test[i:i+(48+4)]
    to_predict = section[-4:]
    to_predict.iloc[:, 1:] = section[47:48].iloc[:, 1:]
    all_predictions.append(to_predict)
    i += (48+4)

preds = pd.concat(all_predictions)

preds.to_csv("closest_time.csv", index=False)

```
